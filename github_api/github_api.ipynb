# Reading from repository

## Setting up

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import os 

github_token = os.environ['GITHUB_TOKEN']
github_owner = os.environ['GITHUB_OWNER']
github_repo = os.environ['GITHUB_REPO']
headers = {
    'Authorization': f'Bearer {github_token}',
    'Accept': 'application/vnd.github+json'
}
url = f'https://api.github.com/repos/{github_owner}/{github_repo}/git/trees/main?recursive=true'
```

## Request to API

```python
import requests

response = requests.get(url=url, headers=headers)
if response.ok:
    items = response.json()['tree']
    for item in items:
        if item['type'] == 'blob' and item['url']:
            cont_response = requests.get(url=item['url'], headers=headers)
            if cont_response.ok:
                print(cont_response.json()['content'])
else:
    print(">>> API error!")
```
